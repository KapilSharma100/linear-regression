```python
import pandas as pd
```

```python
data=pd.read_csv('Salary_Data.csv')
print(data)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X=data.iloc[:,0]
Y=data.iloc[:,1]
plt.scatter(X,Y)
plt.xlabel('X-axis')
plt.ylabel('y-axis')
plt.title('data set scatter plot')
plt.show()
```

```python
mean_x=np.mean(X)
mean_y=np.mean(Y)
n=len(X)
num,den=0,0
for i in range(n):
    num+=((X[i]-mean_x)*(Y[i]-mean_y))
    den+=(X[i]-mean_x)**2
m=num/den
c=mean_y-(m*mean_x)
print(m)
print(c)
```

```python
max_x=np.max(X)+5
min_x=np.min(X)-5
x=np.linspace(min_x,max_x,100)
y=c+m*x
plt.plot(x,y,color='#52b920',label='regression line')
plt.scatter(X,Y,color='blue',label='scatter plot')
plt.xlabel('X-axis')
plt.ylabel('y-axis')
plt.title('final plot')
plt.legend()
plt.show()
```
